# Datos HECHT

En este notebook vamos a preparar los datos proporcionados por HECHT para la detección de comunidades. Los atributos de cada uno de los registros se ecuentran explicados en el Excel almacenado en la carpeta `/data/`.


```python
hecht_df = pd.read_csv('../../data/HECHT/dataset_15122021.csv', sep=';')
hecht_df.head()
```

Analizamos cada uno de los atributos para ver los posibles valores que tiene o si están incompletos.

```python
def print_data(df, name_attribute, num_registers):
    print(name_attribute)
    val = hecht_df.groupby(by=name_attribute).count().iloc[:,0]

    print('Num. Values:', val.values.sum(), '/', num_registers)
    print('Values:')
    for i in val.index:
        print('\t-', i, val[i])

    print('------------')
```

```python
attributes = hecht_df.columns.values
num_registers = len(hecht_df)
```

## Atributos demográficos

En primer lugar, analizaremos los atributos demográficos. Estos atributos son los siguientes:
- DemographicEducationType
- DemographicGender
- DemographicGrade
- DemographicIdentity
- DemographicPolitics
- DemographicReligous
- DemographicsPrepDuration

```python
list_attributes = ['DemographicEducationType', 'DemographicGender', 'DemographicGrade', 'DemographicIdentity', 'DemographicPolitics','DemographicReligous', 'DemographicsPrepDuration']

for name in list_attributes:
    print_data(hecht_df, name, num_registers)
```

Estos atributos los podemos utilizar para definir las comunidades explícitas (por ejemplo usuarios arabe-cristianos (AC)). El único atributo que eliminamos de aquí es el correspondiente a `DemographicsPrepDuration` que mide el tiempo que tardaron los usuarios en completar el test demográfico.

```python
hecht_df.drop(columns='DemographicsPrepDuration', inplace=True)
```
