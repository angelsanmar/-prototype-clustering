```python
import pandas as pd

from context import community_module
from community_module.community_detection.similarityCommunityDetection import SimilarityCommunityDetection
from community_module.similarity.emotionSimilarity import EmotionSimilarity
```

```python
# Paso 1, Cargamos los datos
users_mncn = pd.read_csv('../data/MNCN/user_profiles.csv')
users = users_mncn.iloc[:,1:]
```

```python
# Paso 2, Detectamos comunidades
community_detection = SimilarityCommunityDetection(users)
result = community_detection.calculate_communities(metric='cosine', n_clusters=4)
```

```python
# Agrego los grupos a los usuarios
users_mncn['group'] = result.values()
```
